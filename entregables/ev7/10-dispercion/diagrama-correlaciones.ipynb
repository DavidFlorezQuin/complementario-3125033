```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('datos-programa.csv')
```

```python
correlation_matrix = data.select_dtypes(include=['float64', 'int64']).corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap de Correlación entre Métricas de Programación')
plt.show()
```

```python
data = pd.read_csv('datos-programa.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data,
    x='ExperienciaAnios',
    y='SalarioMensualUSD',
    hue='Lenguaje',
    style='Lenguaje',
    palette='tab10',
    s=100
)

# configurar el gráfico
plt.title('Relación entre Experiencia y Salario por Lenguaje de Programación', fontsize=16)
plt.xlabel('Años de Experiencia', fontsize=12)
plt.ylabel('Salario Mensual (USD)', fontsize=12)
plt.legend(title='Lenguaje', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Configurar el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Diagrama de dispersión con línea de tendencia
sns.scatterplot(
    data=data,
    x='ExperienciaAnios',
    y='SalarioMensualUSD',
    hue='Lenguaje',
    style='Lenguaje',
    palette='tab10',
    s=100
)

# ajuste de la linea

sns.regplot(
    data=data,
    x='ExperienciaAnios',
    y='SalarioMensualUSD',
    scatter=False,
    color='black',
    line_kws={'linestyle': 'dashed'}
)

# Personalizar el gráfico
plt.title('A mayor años de experiencia, mayor será el salario de la persona', fontsize=16)
plt.xlabel('Años de Experiencia', fontsize=12)
plt.ylabel('Salario Mensual (USD)', fontsize=12)
plt.legend(title='Lenguaje', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
